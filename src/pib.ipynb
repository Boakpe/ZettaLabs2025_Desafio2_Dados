```python
import pandas as pd
```

### Filtra o dados para apenas as cidade do estado do Pará

```python
pib_2002_2009 = pd.read_excel('../data/PIB/PIB dos Municípios - base de dados 2002-2009.xls')
pib_2010_2021 = pd.read_excel('../data/PIB/PIB dos Municípios - base de dados 2010-2021.xlsx')
```

```python
somente_para = lambda x: x['Sigla da Unidade da Federação'] == 'PA'
pib_2002_2009_pa = pib_2002_2009[somente_para(pib_2002_2009)]
pib_2010_2021_pa = pib_2010_2021[somente_para(pib_2010_2021)]
```

```python
# Salva os arquivos filtrados
pib_2002_2009_pa.to_csv('../data/PIB/PIB_2002_2009_PA.csv', index=False, encoding='utf-8')
pib_2010_2021_pa.to_csv('../data/PIB/PIB_2010_2021_PA.csv', index=False, encoding='utf-8')
```

### Remove as colunas que não tem dados que variam

```python
def remove_constant_columns(df):
    return df.loc[:, (df != df.iloc[0]).any()]

pib_2002_2009_pa = remove_constant_columns(pib_2002_2009_pa.copy())
pib_2010_2021_pa = remove_constant_columns(pib_2010_2021_pa.copy())
```

### Seleciona apenas as colunas que tem dados que importam para o projeto

```python
pib_2002_2009_pa_filtered = pib_2002_2009_pa[
    [
        "Ano",
        "Código do Município",
        "Nome do Município",
        "Produto Interno Bruto per capita, \na preços correntes\n(R$ 1,00)",
        "Produto Interno Bruto, \na preços correntes\n(R$ 1.000)",
        "Valor adicionado bruto da Agropecuária, \na preços correntes\n(R$ 1.000)",
        "Valor adicionado bruto da Indústria,\na preços correntes\n(R$ 1.000)",
        "Valor adicionado bruto dos Serviços,\na preços correntes \n- exceto Administração, defesa, educação e saúde públicas e seguridade social\n(R$ 1.000)" 
    ]
].copy()
```

```python
pib_2010_2021_pa_filtered = pib_2010_2021_pa[
    [
        "Ano",
        "Código do Município",
        "Nome do Município",
        "Produto Interno Bruto per capita, \na preços correntes\n(R$ 1,00)",
        "Produto Interno Bruto, \na preços correntes\n(R$ 1.000)",
        "Valor adicionado bruto da Agropecuária, \na preços correntes\n(R$ 1.000)",
        "Valor adicionado bruto da Indústria,\na preços correntes\n(R$ 1.000)",
        "Valor adicionado bruto dos Serviços,\na preços correntes \n- exceto Administração, defesa, educação e saúde públicas e seguridade social\n(R$ 1.000)" 
    ]
].copy()
```

### Junta os dois dataframes

```python
pib_2002_2021_pa = pd.concat([pib_2002_2009_pa_filtered, pib_2010_2021_pa_filtered], ignore_index=True)
```

```python
pib_2002_2021_pa.sort_values(by=['Nome do Município', 'Ano'])
```

```python
pib_2002_2021_pa.to_csv('../data/PIB/PIB_2002_2021_PA.csv', index=False, encoding='utf-8')
```
